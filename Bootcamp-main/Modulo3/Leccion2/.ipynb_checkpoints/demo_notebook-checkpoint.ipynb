```python
import pandas as pd

csv_path = "Titanic-Dataset.csv"
```

```python
df = pd.read_csv(csv_path)
```

```python
df.head()
```

<h3>1) Selección con .loc y .iloc</h3>

```python
df.iloc[0:3, 0:4]
```

```python
df.loc[0:2, ["Survived", "Sex", "Age"]]
```

<h3>2) Filtros con & y |</h3>

```python
filtro_and = (df["Sex"] == "female") & (df["Pclass"] == 1)
```

```python
df.loc[filtro_and, ["Survived", "Pclass", "Sex", "Age", "Fare"]].head()
```

```python
filtro_or = (df["Pclass"] == 1) | (df["Survived"] == 1)
```

```python
df.loc[filtro_or, ["Survived", "Pclass", "Sex", "Age", "Fare"]].head()
```
